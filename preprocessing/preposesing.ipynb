```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', ' ', text) 
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
import pandas as pd

def hitung_jumlah_kata(dataset_list, dataset_names):
    """
    Menghitung jumlah kata total dan rata-rata per dataset.
    
    Params:
    - dataset_list: list of DataFrame, misalnya [ugm, usk, mojok, tempo, kompas]
    - dataset_names: list of str, misalnya ['ugm', 'usk', 'mojok', 'tempo', 'kompas']
    
    Return:
    - DataFrame ringkasan jumlah kata
    """
    hasil = []

    for name, df in zip(dataset_names, dataset_list):
        # pastikan kolom tersedia
        if 'konten_tokens' not in df.columns or 'judul_tokens' not in df.columns:
            raise ValueError(f"Dataset {name} tidak memiliki kolom konten_tokens atau judul_tokens")
        
        # hitung jumlah kata di tiap baris
        df['konten_len'] = df['konten_tokens'].apply(lambda x: len(x) if isinstance(x, list) else 0)
        df['judul_len'] = df['judul_tokens'].apply(lambda x: len(x) if isinstance(x, list) else 0)
        
        total_konten = df['konten_len'].sum()
        total_judul = df['judul_len'].sum()
        total_semua = total_konten + total_judul
        
        rata2_konten = df['konten_len'].mean()
        rata2_judul = df['judul_len'].mean()

        hasil.append({
            'dataset': name,
            'total_konten_words': total_konten,
            'total_judul_words': total_judul,
            'total_semua_words': total_semua,
            'rata2_konten_per_dokumen': rata2_konten,
            'rata2_judul_per_dokumen': rata2_judul,
            'jumlah_dokumen': len(df)
        })

    return pd.DataFrame(hasil)
```

```python
dataset_clean = [ugm, usk, mojok, tempo, kompas]
dataset_names = ['UGM', 'USK', 'Mojok', 'Tempo', 'Kompas']

hasil_kata = hitung_jumlah_kata(dataset_clean, dataset_names)
print(hasil_kata)
```

```python
import pandas as pd

def combine_datasets_keep_cols(dataset_list, dataset_names,
                               cols_tokens=('konten_tokens','judul_tokens'),
                               cols_len=('konten_len','judul_len'),
                               source_col='source'):
    """
    Gabungkan beberapa DataFrame dan hasilkan DataFrame yang hanya berisi:
      konten_tokens, judul_tokens, konten_len, judul_len, source

    Params:
      dataset_list : list of DataFrame [ugm, usk, ...]
      dataset_names: list of nama sumber ['UGM','USK',...]
    Return:
      combined_df : pandas DataFrame gabungan dengan kolom yang diminta
    """
    if len(dataset_list) != len(dataset_names):
        raise ValueError("dataset_list dan dataset_names harus sama panjang")

    dfs = []
    for df, name in zip(dataset_list, dataset_names):
        df_copy = df.copy()

        # pastikan kolom token ada; jika belum, buat kolom kosong atau tokenisasi
        for tcol in cols_tokens:
            if tcol not in df_copy.columns:
                # jika tidak ketemu, buat kolom kosong list agar tidak error
                df_copy[tcol] = [[] for _ in range(len(df_copy))]

        # pastikan kolom panjang ada; jika tidak, hitung dari token list
        if cols_len[0] not in df_copy.columns:
            df_copy[cols_len[0]] = df_copy[cols_tokens[0]].apply(lambda x: len(x) if isinstance(x, list) else 0)
        if cols_len[1] not in df_copy.columns:
            df_copy[cols_len[1]] = df_copy[cols_tokens[1]].apply(lambda x: len(x) if isinstance(x, list) else 0)

        # tambahkan kolom source (label)
        df_copy[source_col] = name

        # pilih hanya kolom yang diperlukan
        keep_cols = [cols_tokens[0], cols_tokens[1], cols_len[0], cols_len[1], source_col]
        # pastikan kolom urut
        df_copy = df_copy.reindex(columns=keep_cols)

        dfs.append(df_copy)

    # gabungkan semuanya
    combined_df = pd.concat(dfs, ignore_index=True, sort=False)
    return combined_df

# =======================
# Contoh pemakaian
# =======================
dataset_clean = [ugm, usk, mojok, tempo, kompas]
dataset_names = ['UGM', 'USK', 'Mojok', 'Tempo', 'Kompas']

combined = combine_datasets_keep_cols(dataset_clean, dataset_names)
print(combined.shape)
print(combined.head())
```

```python
import os
import pickle
from tqdm import tqdm
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stem_tokens_list(tokens, stemmer):
    """Stem satu list token; jika bukan list, kembalikan list kosong."""
    if not isinstance(tokens, list):
        return []
    return [stemmer.stem(t) for t in tokens]

def stem_dataframe_in_chunks(df,
                             konten_col='konten_tokens',
                             judul_col='judul_tokens',
                             chunk_size=5000,
                             save_path=None,
                             inplace=True,
                             start_chunk=0):
   
    # init stemmer factory (bikin per proses agar thread-safe)
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    
    n = len(df)
    n_chunks = (n + chunk_size - 1) // chunk_size
    os.makedirs(save_path, exist_ok=True) if save_path else None

    target_cols = []
    if inplace:
        target_cols = [konten_col, judul_col]
    else:
        target_cols = [f"{konten_col}_stem", f"{judul_col}_stem"]
        # init new cols empty
        df[target_cols[0]] = [[] for _ in range(n)]
        df[target_cols[1]] = [[] for _ in range(n)]

    for chunk_idx in tqdm(range(start_chunk, n_chunks), desc="Stemming chunks"):
        i0 = chunk_idx * chunk_size
        i1 = min(n, (chunk_idx + 1) * chunk_size)
        # slice view
        slice_idx = range(i0, i1)

        # apply stemming on konten
        for col, out_col in [(konten_col, target_cols[0]), (judul_col, target_cols[1])]:
            # process rows in slice
            # local stemmer instance used already (Sastrawi Stemmer object reused safely here)
            for ridx in slice_idx:
                toks = df.at[ridx, col] if col in df.columns else []
                # guard for non-list
                if not isinstance(toks, list):
                    res = []
                else:
                    # faster list comprehension
                    res = [stemmer.stem(t) for t in toks]
                df.at[ridx, out_col] = res

        if save_path:
            fname = os.path.join(save_path, f"stem_checkpoint_chunk_{chunk_idx:04d}.pkl")
            with open(fname, 'wb') as f:
                pickle.dump({'chunk_idx': chunk_idx, 'i0': i0, 'i1': i1, 'df_slice': df.iloc[i0:i1]}, f)
    return df
```

```python

combined_df = stem_dataframe_in_chunks(
    combined,
    konten_col='konten_tokens',
    judul_col='judul_tokens',
    chunk_size=1000,            
    save_path='checkpoints_stem',
    inplace=True
)
```

```python
import os
import re

checkpoint_dir = 'checkpoints_stem'
if not os.path.isdir(checkpoint_dir):
    print(f"Directory not found: {checkpoint_dir}")
else:
    files = []
    for fn in os.listdir(checkpoint_dir):
        m = re.match(r'stem_checkpoint_chunk_(\d{4})\.pkl$', fn)
        if m:
            files.append((int(m.group(1)), os.path.join(checkpoint_dir, fn)))
    files.sort()

    dfs = []
    for idx, fpath in files:
        try:
            df_slice = pd.read_pickle(fpath)['df_slice']
            dfs.append(df_slice)
            print(f"Loaded chunk {idx:04d}: shape {df_slice.shape}")
        except Exception as e:
            print(f"Failed to load {fpath}: {e}")

    if dfs:
        df_all = pd.concat(dfs, ignore_index=True, sort=False)
    else:
        df_all = pd.DataFrame()

    print("Combined dataframe shape:", df_all.shape)
    print(df_all.head())
```
